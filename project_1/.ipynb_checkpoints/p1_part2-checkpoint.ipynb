*Import Cleaned Data*

```python
import pandas as pd
import numpy as np #importing numpy
cars = pd.read_csv('../project_data/project1_data.csv')
# Replace '?' with NaN in the entire DataFrame
cars.replace('?', pd.NA, inplace=True)
# Drop rows with NaN values in any column
cars.dropna(inplace=True)
cars_numeric_cols = cars.select_dtypes(include=[np.number])
cars_one_hot_encoded = pd.get_dummies(cars_numeric_cols, columns=['origin'], prefix='origin')
cars_cleaned = cars_one_hot_encoded
```

```python
cars_cleaned.head
```

```python
# Create a array of arrays for weight and mpg
dataset_mpg = [[mpg] for mpg in cars_cleaned['mpg']]
dataset_mpg
```

```python
# Drop the 'mpg' column to create the features DataFrame
X = cars_cleaned.drop('mpg', axis=1).values  # independent variables
y = dataset_mpg               # dependent variable 'mpg'
```

```python
X
```

```python
# Split dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Testing the accuracy
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Predict mpg on the testing data
y_pred = model.predict(X_test)  # Changed variable name here to y_pred for consistency

# Calculate the performance metrics
mse = mean_squared_error(y_test, y_pred)  # Using y_pred now
r2 = r2_score(y_test, y_pred)  # Using y_pred now

# Print out the metrics
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared: {r2}')
```

```python
# For training data R^2 score
train_r2 = model.score(X_train, y_train)
print(f'Training Data R-squared: {train_r2}')

# For testing data R^2 score
test_r2 = model.score(X_test, y_test)
print(f'Testing Data R-squared: {test_r2}')
```
