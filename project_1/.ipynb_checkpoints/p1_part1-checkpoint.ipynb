```python
# Replace '?' with NaN in the entire DataFrame
cars.replace('?', pd.NA, inplace=True)

# Drop rows with NaN values in any column
cars.dropna(inplace=True)

cars.info()
```

**deriving statistical information from the data**

```python
cars.describe()
```

```python
cars_numeric_cols = cars.select_dtypes(include=[np.number])  # taking out non-numerica columns


# Identify potential outliers using IQR (Interquartile Range) for each column
Q1 = cars_numeric_cols.quantile(0.25)
Q3 = cars_numeric_cols.quantile(0.75)
IQR = Q3 - Q1

# Define outliers as those beyond 1.5 * IQR from the Q1 and Q3
outliers = ((cars_numeric_cols < (Q1 - 1.5 * IQR)) | (cars_numeric_cols > (Q3 + 1.5 * IQR))).sum()

outliers
```

*there seems to be 1 mpg outlier and 7 outliers in acceleration*

```python
# One-hot encode the 'origin' column
cars_one_hot_encoded = pd.get_dummies(cars_numeric_cols, columns=['origin'], prefix='origin')
```

```python
cars_one_hot_encoded.info()
```

**univariate analysis and bivariate analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Univariate Analysis Plots
# Histogram for MPG
plt.figure(figsize=(10, 6))
sns.histplot(cars_one_hot_encoded['mpg'], kde=True, bins=20, color='skyblue')
plt.title('Distribution of MPG')
plt.xlabel('Miles Per Gallon (MPG)')
plt.ylabel('Frequency')
plt.show()

# Bivariate Analysis Plots
# Scatter plot for MPG vs Weight
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cars_one_hot_encoded, x='weight', y='mpg', hue='cylinders', palette='coolwarm', alpha=0.6)
plt.title('MPG vs Weight by Cylinder Count')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.legend(title='Cylinders')
plt.show()

# Box plot for MPG by Cylinders
plt.figure(figsize=(10, 6))
sns.boxplot(data=cars_one_hot_encoded, x='cylinders', y='mpg', palette='Set2')
plt.title('MPG Distribution by Cylinder Count')
plt.xlabel('Cylinders')
plt.ylabel('MPG')
plt.show()

# Pairplot for selected variables
selected_columns = ['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']
sns.pairplot(cars_one_hot_encoded[selected_columns], corner=True, plot_kws={'alpha':0.6}, diag_kws={'bins':20})
plt.suptitle('Pairplot of Selected Variables', y=1.02)  # Adjust title position
plt.show()
```

**find correlations between columns**

```python
# Calculate the correlation matrix for the numeric columns in the dataset
correlation_matrix = cars_one_hot_encoded.corr()

# Display the correlation matrix
correlation_matrix
```

```python
# Display the correlation matrix using a heatmap for better visualization
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix with Heatmap')
plt.show()
```

*MPG and weight: There is a very strong negative correlation of -0.831741, suggesting that as the weight of a vehicle increases, its fuel efficiency (MPG) tends to decrease significantly.

MPG and displacement: Similarly, there is a strong negative correlation of -0.804203 between MPG and displacement, indicating that vehicles with larger engines tend to have lower fuel efficiency.

MPG and cylinders: The negative correlation of -0.775396 shows that cars with more cylinders tend to have lower MPG, which is consistent with the understanding that more cylinders generally mean larger engines and hence lower fuel efficiency.*
