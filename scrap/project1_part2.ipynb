```python
import pandas as pd
import numpy as np #importing numpy
cars = pd.read_csv('../project_data/project1_data.csv')
```

```python
cars_numeric_cols = cars.select_dtypes(include=[np.number])
cars_one_hot_encoded = pd.get_dummies(cars_numeric_cols, columns=['origin'], prefix='origin')
```

```python
cars_cleaned = cars_one_hot_encoded
```

```python
cars_cleaned.info()
```

```python
cars_cleaned
```

```python
import sklearn
from sklearn import linear_model
lr = linear_model.LinearRegression()

# Split into training and testing 

# Define the split point
split_point = 200

# Split the DataFrame into training and testing sets
training_data = cars_cleaned.iloc[:split_point]
testing_data = cars_cleaned.iloc[split_point:]
```

```python
training_data
```

```python
testing_data
```

```python
# Assuming cars_cleaned is a DataFrame with the columns 'weight' and 'mpg'
training_dataset_weight = [[weight] for weight in training_data['weight']]
training_dataset_mpg = [[mpg] for mpg in training_data['mpg']]
testing_dataset_weight = [[weight] for weight in testing_data['weight']]
testing_dataset_mpg = [[mpg] for mpg in testing_data['mpg']]
```

```python
training_dataset_weight
```

```python
training_dataset_mpg
```

```python
lr.fit(training_dataset_weight, training_dataset_mpg)
```

```python
test_predict_weight = lr.predict(test_dataset_weight) # values predicted by model on the test data
```
